```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
```

```python
x=np.arange(0,100,5)
print(x)
print(len(x))
```

```python
x1=np.arange(0,100,4)
y= [i for i in range(len(x))]
y1= [i for i in range(len(x1))]
```

```python
np.random.random()
```

```python
plt.xlabel("random x")
plt.ylabel("random y")
plt.xlim(0,100) #here we are setting the limit of x and y
plt.ylim(0,30)

plt.plot(x,y,label="numpy data view")
plt.plot(x1,y1,label="numpy data1 view")
plt.bar(x,y,label="numpy data view in bar")
plt.bar(x1,y1,label="numpy data1 view in bar")
plt.legend()
plt.grid(color='g')
plt.show()
```

```python

players=['vk','rohit','msd','jadeja']
runs=[1000,800,700,500]
plt.xlabel("players")
plt.ylabel("score")
plt.ylim(400,1300) #here we are setting the limit of x and y
plt.bar(players,runs,label="performance",color='b')
plt.legend()

plt.grid(color='g')
plt.show()
```

```python
players=['vk','rohit','msd','jadeja']
run=[1200,800,700,500]
run1=[1100,860,770,500]

fig,ax=plt.subplots(2)
ax[0].bar(players,run,label="2019",color="g")
ax[0].legend()
ax[1].bar(players,run1,label='2020')
ax[1].legend()
plt.show()
```

```python
players=['vk','rohit','msd','jadeja']
run=[1200,800,700,500]
plt.pie(run,labels=players,shadow=True,autopct='%1.1f%%')# %1.1f% is formula to calulate the percentage from given data
plt.legend()
plt.show()
```

```python
players=['vk','rohit','msd','jadeja']
run=[1200,800,700,500]
plt.pie(run,labels=players,shadow=1,autopct='%1.1f%%',explode=(0.3,0,0,0))# %1.1f% is formula to calulate the percentage from given data
plt.legend()
plt.show()
```

```python
players=['vk','rohit','msd','jadeja']
run=[1200,800,700,500]
run1=[1230,900,600,400]
fig,ax=plt.subplots(2)
ax[0].pie(run,labels=players,shadow=1,autopct='%1.1f%%',explode=(0.3,0,0,0))# %1.1f% is formula to calulate the percentage from given data
ax[1].pie(run1,labels=players,shadow=1,autopct='%1.1f%%',explode=(0,0.2,0,0))
ax[0].legend()
ax[1].legend()
plt.show()
```

```python
plt.scatter(x,y,label="star way",marker='*',s=80)
plt.scatter(x1,y1,label="Triangle way",marker="^",s=90)#scatter formate
plt.grid(color='g')
plt.legend()
plt.show()
```
